```python
# Flask 웹 서버
from flask import Flask, request, send_file

from datetime import datetime

app = Flask(__name__)


@app.route("/")
def home():

    ip = request.remote_addr

    time = datetime.now()

    user_agent = request.headers.get("User-Agent")

    print("========== 접속 발생 ==========")
    print("접속 시간:", time)
    print("접속 IP:", ip)
    print("브라우저 정보:", user_agent)

    return """
    <html>

    <head>
        <title>트래픽 수집 테스트</title>
    </head>

    <body>

        <h1>트래픽 수집 테스트 웹사이트</h1>

        <p>
        이 웹사이트는 접속 정보와 영상 트래픽을 수집합니다.
        </p>

        <hr>

        <h2>영상 재생 테스트</h2>

        <!-- 영상 재생 영역 -->
        <!-- 사용자가 영상을 재생하면 /video 경로로 요청이 발생함 -->
        <video width="640" controls>
            <source src="/video" type="video/mp4">
            브라우저가 video 태그를 지원하지 않습니다.
        </video>

        <br><br>

        <!-- 새로고침 버튼 -->
        <!-- 버튼을 누르면 현재 페이지가 다시 로드되어 접속 로그가 다시 출력됨 -->
        <button onclick="location.reload()">
            새로고침 테스트
        </button>

    </body>

    </html>
    """


@app.route("/video")
def video():

    ip = request.remote_addr

    time = datetime.now()

    print("========== 영상 요청 발생 ==========")
    print("영상 요청 시간:", time)
    print("영상 요청 IP:", ip)


    return send_file(
        "sample.mp4",
        mimetype="video/mp4"
    )


# 서버 실행 부분
if __name__ == "__main__":

    app.run(
        host="0.0.0.0",
        port=5000
    )
```
