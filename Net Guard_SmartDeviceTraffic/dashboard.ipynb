```python

import pandas as pd

import streamlit as st

st.title("스마트폰 트래픽 분석 대시보드")

df = pd.read_csv("traffic_log.csv")

df["time"] = pd.to_datetime(df["time"])

df["data_size_mb"] = df["data_size_bytes"] / 1024 / 1024

df["minute"] = df["time"].dt.strftime("%H:%M")

st.subheader("전체 로그")
st.dataframe(df)

ip_list = df["ip"].unique()

selected_ip = st.selectbox("기기 IP 선택", ip_list)

ip_df = df[df["ip"] == selected_ip]

start_time = ip_df["time"].min()
end_time = ip_df["time"].max()

usage_minutes = (end_time - start_time).total_seconds() / 60

total_mb = ip_df["data_size_mb"].sum()

video_count = len(ip_df[ip_df["request_type"] == "VIDEO"])

st.subheader("요약 분석")

st.metric("사용 시간", f"{usage_minutes:.1f}분")

st.metric("총 데이터 사용량", f"{total_mb:.2f} MB")

st.metric("영상 요청 횟수", f"{video_count}회")

st.subheader("1분 단위 데이터 사용량")

minute_usage = ip_df.groupby("minute")["data_size_mb"].sum()

st.line_chart(minute_usage)

st.subheader("1분 단위 요청 횟수")

minute_request_count = ip_df.groupby("minute").size()

st.bar_chart(minute_request_count)

st.subheader("요청 종류별 횟수")

request_count = ip_df["request_type"].value_counts()

st.bar_chart(request_count)
```
